```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('data.csv')
```

```python
df
```

```python
df.info()
```

```python
gen_df=pd.get_dummies(df['Gender'],drop_first=True)
gen_df
```

```python
df.drop('Gender',axis=1,inplace=True)
```

```python
df=pd.concat([df,gen_df],axis=1)
```

```python
df
```

```python
X=df.drop('Purchased_moble',axis=1)
X
```

```python
y=df.Purchased_moble
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_predict=classifier.predict(X_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
ac=accuracy_score(y_test,y_predict)
ac
```

```python
import pickle
pickle.dump(sc,open('mobile_pursc.pkl','wb'))
pickle.dump(classifier,open('scaler.pickle','wb'))
```
